```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

Since I have written %matplotlib inline that is way I need not to write plt.show()

```python
y=x*x*(-1)
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,'r')
plt.subplot(2,2,3)
plt.plot(x,y,'g+')
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
x=np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,'r')
y = np.cos(x)
plt.subplot(2,2,3)
plt.plot(x,y,'g+')
plt.subplot(2,2,4)
plt.plot(x,y,'g*')
```
